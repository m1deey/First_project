```python
import pandas as pd
df=pd.read_csv('train.csv')
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df["Sex"]=df["Sex"].map({"male":1,"female":0})
df
```

```python
df["Age"]=df["Age"].fillna(df["Age"].median())
df.isnull().sum()
```

```python
df["Cabin"] = df["Cabin"].notnull().astype(int)
```

```python
#df=df.drop(columns="Ticket")
df=df.drop(columns="PassengerId")
```

```python
df["Title"] = df["Name"].str.split(",").str[1].str.split(".").str[0].str.strip()
df
```

```python
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
df=df.drop(columns="Name")
df=pd.get_dummies(df,columns=["Embarked"],dtype=int)
```

```python
df
```

```python
bins = [0, 12, 18, 25, 35, 50, 65, 80]
labels = ["Kid", "Teen", "Young Adult", "Adult", "Mid Age", "Senior", "Elder"]

df["AgeGroup"] = pd.cut(df["Age"], bins=bins, labels=labels)
df
```

```python
df["FareGroup"] = pd.qcut(df["Fare"], q=4, labels=["Cheap", "Normal", "Expensive", "Luxury"])
```

```python
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1
df["IsAlone"] = (df["FamilySize"] == 1).astype(int)
```

```python
import matplotlib.pyplot as plt
corr=df.corr(numeric_only=True)
plt.figure(figsize=(10,8))
plt.imshow(corr,cmap="coolwarm")
plt.colorbar()
plt.xticks(range(len(corr.columns)),corr.columns,rotation=90)
plt.yticks(range(len(corr.columns)),corr.columns)
plt.title("correlation heatmap")
plt.tight_layout()
plt.show()
```

```python
sr=df["Survived"].value_counts().rename(index={0:"Didn't survive",1:"Survived"})
```

```python
plt.figure(figsize=(10,8))
plt.bar(sr.index,sr.values,color=["red","green"],edgecolor="black")
plt.title("Survival count",fontsize=40,fontweight="bold")
plt.show()
```

```python
group=df.groupby("Sex")["Survived"].mean().rename(index={0:"female",1:"male"})
```

```python
plt.figure(figsize=(10,8))
plt.bar(group.index,group.values,color=["pink","blue"],edgecolor="black")
plt.xlabel("sex",fontsize=30)
plt.ylabel("survived",fontsize=30)
plt.title("Survival by sex",fontsize=40,fontweight="bold")
plt.show()
```

```python
ss=df.groupby("Pclass")["Survived"].mean()
plt.figure(figsize=(10,8))
plt.bar(ss.index,ss.values,color=["purple","yellow","orange"],edgecolor="black")
plt.xlabel("Pclass",fontsize=30)
plt.ylabel("survived",fontsize=30)
plt.title("Survival by Pclass",fontsize=40,fontweight="bold")
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.hist(df["Age"],bins=30,edgecolor="black",color="skyblue")
plt.xlabel("Ages",fontsize=30)
plt.ylabel("Frequency",fontsize=30)
plt.title("Age distribution")
plt.show()plt.figure(figsize=(10,8))
```

```python
gp=df.groupby("AgeGroup")["Survived"].mean()
plt.figure(figsize=(10,8))
plt.barh(gp.index,gp.values,edgecolor="black")
plt.xlabel("Survived",fontsize=30)
plt.ylabel("AgesGroup",fontsize=30)
plt.title("AgeGroup distribution")
plt.show()
```

```python
fr=df.groupby("FareGroup")["Survived"].mean()
plt.figure(figsize=(10,8))
plt.barh(fr.index,fr.values,edgecolor="black",color="green")
plt.xlabel("Survived",fontsize=30)
plt.ylabel("FareGroup",fontsize=30)
plt.title("FareGroup distribution")
plt.show()
```

```python
ll=df.groupby("IsAlone")["Survived"].mean().rename(index={0:"NotAlone",1:"Alone"})
plt.figure(figsize=(10,8))
plt.bar(ll.index,ll.values,color=["black","grey"],edgecolor="black")
plt.xlabel("IsAlone",fontsize=30)
plt.ylabel("survived",fontsize=30)
plt.title("Survival by Loneliness",fontsize=40,fontweight="bold")
plt.show()
```

```python
plt.figure(figsize=(10,8))
fam = df.groupby("FamilySize")["Survived"].mean()
plt.plot(fam.index, fam.values, marker="o", color="teal")
plt.xlabel("Family Size", fontsize=20)
plt.ylabel("Survival Rate", fontsize=20)
plt.title("Survival by Family Size", fontsize=30)
plt.show()
```
